```python
import numpy as np

data = np.genfromtxt('../data/housing.csv', delimiter=None, filling_values=0)
X = data[:,:-1]
Y = data[:, -1]
n_samples = X.shape[0]
W = np.random.rand(X.shape[1], 1)
B = np.random.rand(1)
y_pred = X @ W + B
error = y_pred - Y
grad_w = 2/n_samples * (X.T  @ error)
grad_b = 2/n_samples * np.sum(error)

print(f"X Shape: {X.shape}")
print(f"Y Shape: {Y.shape}")
print(f"W Shape: {W.shape}")
print(f"B Shape: {B.shape}")
print(f"Prediction Shape: {y_pred.shape}")
print(f"Error Shape {error.shape}")
print(f"W gradient shape: {grad_w.shape}")
print(f"B gradient shape: {grad_b.shape}")


# print(X)
# print(Y)
# print(W)
print(y_pred)
print(error)
print(grad_w)
print(grad_b)
```
